```python
import requests
import pandas as pd
import numpy as np
import json
import time
from datetime import datetime

# ============================================================
# KONFIGURASI
# ============================================================
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "cookie": "t_fv=1775100094336; t_uid=BFsRJHj3ABk6f5kkGAIaZYhrDE51nAfV; hng=ID|id-ID|IDR|360; hng.sig=dJwrVwSueShOlZz95EBCvlH9FLAVtzGZ3msUnc25HIQ; lwrid=AgGdTDY%2F8Mw27vD5MBUMX38RnJ8X; lzd_cid=c0aa823f-c1e0-4b83-9fcc-7cd0ffe862af; lzd_sid=135433624d7c8e8edd32a229267d5433; _tb_token_=e86ed6b51b5e5; xlly_s=1; lwrtk=AAIEac5RYn7ksV6ewl55W0OZy5Y0t954Hoj2dYgTGt2TDsIeYpAM0RE=; cna=6s5UIpphtiwCAXIFWbgHV0+w; _m_h5_tk=64f02424eca0dee3421186253533f3e0_1775107698561; _m_h5_tk_enc=cf406db5a918eb6ce027d46979959d68; x5sec=7b22617365727665722d6c617a6164613b33223a22617c434a375974383447454971753864634349676c795a574e686348526a614745776c4f653476775a4b616a41324d47466d5a6a41774d4441774d4441774d4441774d4441774f4441774d4441774d4441774d4441774d4441774d4441774d4451334d6a41314e546c694f4441324d7a6b304d6a6c68595452684e47526c5a4463774d4441774d4441774d4441785a6a45794d6d597a4d32517a5a546b314d3245334d446b78596a5a6c4d4751784e5449314f544a6d4e7a593d222c22733b32223a2237383764396233356161396665653231227d; t_sid=CUZMdrJUAuDSw9L4ehoVNcsqLulPw2Od; utm_origin=https://www.bing.com/; utm_channel=SEO; epssw=12*FQ7i1wwGGI602x5zIGGh7LSHC6_2H0_iDk3EZUXFGGGhMsXF4URzl6h7rrr_esGg4wGGHL46ayTSlrjrjrjVaAAdvCIMT_hpkuGkX8Cyo_3cudf8Xlq3Ko4rO5bKRMEAj7XGbd2Yd4xQyvZcjrw6GsQJGGGGGRbkx2xsjj0Jo3IGGT_-O1BIIlSaC4LBcTpi6Z6vr6UMZQGGskqUNzOzcwOiyIGGhwGGt_ixg33zGiIxMkQRXsudE0zRJr9ssnvHtnPQ7EUybCYFKTp4NVBFu4poBin2CxAAistpicMnCAbG3FbUmTt8tLlx4L2e5Gt.; tfstk=gIxZBs66OcnamdckULjq8iLbtcjOXil73n1fnKvcC1fgflND8Bd4hAhxWe52t696cGcaBedWEltsWiZVXiImNbiS0ApODi0HJczP6pvdnv6goaSwhiImNXiI6O_RDC7dPhQDKvWfHS2Mitb3-6BfisAcswVhUTIciIAc-wXVHoVcSlvht6BcmsAcmvJh9tSciIjmLpcEFZbzYtQiQTxJBmhf6N5kI6rywHWGcPpGToqDYZQNZo1UmoAFIKVZS8Zrl_YR1a1ptDZN0pX2aTRatXSDLhKP-hcL0QRerMWpJjzVaEv6yEbEIcvFb17yANox1G8D6n797-HJKGApy_QiJXBef3_VwwPnsp9Fsa821XEc1UJHTatQOusHEeSF4EFAKh-HDFP0uNXdL_MEL5u65b8kzgoQkrQ_995SIA4YkNXdL_MELrUA52BFNADG."
}

TOKO_LIST = [
    {"nama": "Watsons",             "keyword": "watsons"},
    {"nama": "BeautyHaul",          "keyword": "beauty haul"},
    {"nama": "BeautyHaul Indonesia","keyword": "beautyhaul indonesia"},
]

MAX_HALAMAN = 50   # 1 halaman ≈ 40 produk
DELAY_DETIK = 2
# ============================================================


def scrape_satu_halaman(keyword, halaman):
    url = f"https://www.lazada.co.id/catalog/?ajax=true&q={keyword}&page={halaman}"
    try:
        response = requests.get(url, headers=HEADERS, timeout=10)
        response.raise_for_status()
        data = response.json()
        items = data.get("mods", {}).get("listItems", [])
        return items
    except requests.exceptions.RequestException as e:
        print(f"      ⚠️  Error halaman {halaman}: {e}")
        return []
    except json.JSONDecodeError:
        print(f"      ⚠️  Response bukan JSON di halaman {halaman}")
        return []


def parse_produk(item, nama_toko):
    return {
        "toko"              : nama_toko,
        "nama_produk"       : item.get("name", np.nan),
        "harga"             : item.get("price", np.nan),
        "harga_original"    : item.get("originalPrice", np.nan),
        "diskon"            : item.get("discount", np.nan),
        "rating"            : item.get("ratingScore", np.nan),
        "jumlah_review"     : item.get("review", np.nan),
        "jumlah_terjual"    : item.get("itemSoldCntShow", np.nan),
        "nama_toko_lazada"  : item.get("sellerName", np.nan),
        "lokasi"            : item.get("location", np.nan),
        "kategori"          : item.get("categoryName", np.nan),
        "brand"             : item.get("brandName", np.nan),
        "url_produk"        : "https:" + item.get("itemUrl", "") if item.get("itemUrl") else np.nan,
        "url_gambar"        : item.get("image", np.nan),
        "tanggal_scraping"  : datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    }


def scrape_toko(nama_toko, keyword):
    print(f"\n🔍 Scraping: {nama_toko} (keyword: '{keyword}')")
    semua_produk = []

    for halaman in range(1, MAX_HALAMAN + 1):
        print(f"   📄 Halaman {halaman}/{MAX_HALAMAN}...", end=" ")
        items = scrape_satu_halaman(keyword, halaman)

        if not items:
            print("tidak ada data, berhenti.")
            break

        for item in items:
            semua_produk.append(parse_produk(item, nama_toko))

        print(f"✅ {len(items)} produk")
        time.sleep(DELAY_DETIK)

    print(f"   📦 Total {nama_toko}: {len(semua_produk)} produk")
    return semua_produk


# ============================================================
# JALANKAN SCRAPING
# ============================================================
semua_data = []
for toko in TOKO_LIST:
    data = scrape_toko(toko["nama"], toko["keyword"])
    semua_data.extend(data)

df = pd.DataFrame(semua_data)

# Bersihkan kolom numerik
df["rating"]        = pd.to_numeric(df["rating"], errors="coerce")
df["jumlah_review"] = pd.to_numeric(df["jumlah_review"], errors="coerce")

# Simpan ke Excel
nama_file = f"lazada_produk_{datetime.now().strftime('%Y%m%d_%H%M%S')}.xlsx"
with pd.ExcelWriter(nama_file, engine="openpyxl") as writer:
    df.to_excel(writer, sheet_name="Semua Toko", index=False)
    for toko in df["toko"].unique():
        df[df["toko"] == toko].to_excel(writer, sheet_name=toko, index=False)

print(f"\n✅ Selesai! File tersimpan: {nama_file}")
print(f"📦 Total produk: {len(df)}")
print("\n📋 Preview:")
print(df[["toko", "nama_produk", "harga", "rating"]].head(10).to_string(index=False))
```
